# **t 分布练习 : 利用stats.t 函数**
* 利用stat.t.rvs, 生成一个自由度为2的t分布随机数 ns=20000
* 利用DataFrame.hist画出其直方图；
* 利用t.stats计算其均值，方差
* 利用比较自由度为2，40的直方图；
* 产生两个正态分布，均值和方差分别与上面的t(2),t(40)分布的均值和方差相同，比较这四个分布的直方图，观察一下有什么特点

```python
ns=20000
sampleT1=stats.t.rvs(4,0,1,ns,random_state=100)
sampleT2=stats.t.rvs(40,0,1,ns,random_state=120)
mt1,vt1=stats.t.stats(4,moments="mv")
mt2,vt2=stats.t.stats(40,moments="mv")

sampleN1=stats.norm.rvs(mt1,np.sqrt(vt1),ns,random_state=300)
sampleN2=stats.norm.rvs(mt2,np.sqrt(vt2),ns,random_state=400)

dataSample=pd.DataFrame({"Sample T1":sampleT1,"Sample T2":sampleT2,"Sample N1":sampleN1,"Sample N2":sampleN2})
```

```python
# 步骤1：导入必备库
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
plt.rcParams["font.family"] = ["SimHei", "Microsoft YaHei"]  # 指定中文字体
plt.rcParams["axes.unicode_minus"] = False  # 解决负号显示为方块的问题

# 步骤2：准备绘图数据
x = np.linspace(start=-10, stop=10, num=1000)  # 生成x轴等间距数据
y1 = stats.t.pdf(x, loc=0, df=3, scale=1)      # df=3的t分布概率密度
y2 = stats.t.pdf(x, loc=0, df=30, scale=1)     # df=30的t分布概率密度

# 步骤3：创建画布（自定义大小）
plt.figure(figsize=(8, 5))

# 步骤4：绘制曲线（自定义样式）
plt.plot(x, y1, color="red", linewidth=2, ls="--", label="t分布 (df=3)")
plt.plot(x, y2, color="blue", linewidth=2, ls="-", label="t分布 (df=30)")

# 步骤5：美化标注
plt.title("不同自由度的t分布曲线")
plt.xlabel("x值")
plt.ylabel("概率密度")
plt.legend()  # 显示图例
plt.grid(alpha=0.3)  # 添加网格
plt.ylim(0, 0.4)     # 限定y轴范围

# 步骤6：保存并显示
plt.savefig("t_distribution.png", dpi=300, bbox_inches="tight")
plt.show()
```

# Part 4: 预备知识-LLN, CLT

说明：我们用同一个数据来演示LLN, CLT, 我们产生一个非正态分布的“总体”，然后

（1）LLN：随机采样,观察随着采样大小ns增加的情况下，样本均值的变化； 

（2）CLT 我们以也给固定的ns重复采样，观察样本均值的分布特征


```python
data1= stats.norm.rvs(10,5,5000,random_state=100)
data2= stats.norm.rvs(30,5,2000,random_state=100)
data=np.hstack((data1,data2))
```

```python
plt.hist(data,bins=50)
```

##### 中心极限定理(CLT) Demo:    <font color=#FF0000 >样本均值服从正态分布，均值等于总体均值</font>

```python
sample_means=[]
reps,ns=10000,50
for i in np.arange(10000):
    sample_i=np.random.choice(data,ns,replace=True)
    sample_means=np.append(sample_means,np.mean(sample_i))

print("总体均值是: %.2f " % np.mean(data))
print("样本均值分布的均值是: %.2f " % np.mean(sample_means))

plt.hist(sample_means,bins=50)
plt.axvline(x=np.mean(data), color='red', linestyle='--')
plt.show()
    
```

##### 大数定理（LLN) Demo:样本大小增加，样本均值接近总体均值

```python
sample_means=[]
for ns in np.arange(15,len(data)):
    sample_i=np.random.choice(data,ns,replace=False)
    sample_means=np.append(sample_means,np.mean(sample_i))

print("总体Data的均值为 : %.2f" % np.mean(data))
plt.plot(sample_means)
plt.axhline(y=np.mean(data), color='red', linestyle='--')
plt.show()
    
```

### 数据正态性评估

```python
#偏度（skew）正态分布=0
print(stats.skew(BMEIntro.Scores))
#峰度 正态分布=0
print(stats.kurtosis(BMEIntro.Scores))
#Visually checking the normality
#QQPlot  正态分布：直线拟合

import statsmodels.api as sm
fig = sm.qqplot(BMEIntro.Scores, stats.norm,line='s')
plt.show()

# deensity distribution  # 概率密度分析
BMEIntro.Scores.plot.density(color='green')
```

#### 中心极限定理（CLT) , 样本均值的分布：
- 从这个数据（BMEIntro.Scores）随机采样 重复进行20000次的采样，每次采样150个值（可重复），画出这一万个样本的均值的直方图分布
- 比较数据的均值与样本分布均值

```python
#提示：参考上面例子中的大数定理的演示程序
data=BMEIntro.Scores

sample_means=[]
reps,ns=20000,150
for i in np.arange(reps):
    sample_i=np.random.choice(data, size=ns, replace=True)
    sample_means=np.append(sample_means, sample_i.mean())

print("总体均值是: %.2f " % np.mean(data))
print("样本均值分布的均值是: %.2f " % np.mean(sample_means))

plt.hist(sample_means, bins=50)
fig = sm.qqplot(sample_means, stats.norm,line='s')
plt.show()
```
