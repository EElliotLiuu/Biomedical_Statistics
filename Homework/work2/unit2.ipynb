## 第一题
某班级学生进行了一次数学考试，成绩如下：75，80，90，65，70，85，95，75，80，80。请计算该组数据的均值、中位数、众数、方差、标准差、标准误、变异系数。 (0.5 point)

```python
#加载头文件
import numpy as np
import pandas as pd
import scipy.stats as stats
import statistics as sta
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
#中文显示设置
plt.rcParams["font.family"] = ["SimHei", "Microsoft YaHei"]  
```

```python
#创建数组，储存成绩
math=np.array([75,80,90,65,70,85,95,75,80,80])
#计算各种统计量
print("成绩的均值是：",math.mean())
print("中位数是：",sta.median(math))
print("众数是：",sta.mode(math))
print("方差是：",sta.variance(math))
print("标准差是：",round(sta.stdev(math), 2))
print("标准误是：",round(stats.sem(math), 2))
print("变异系数是：",round(stats.variation(math), 2))
```

## 第二题
把样本进行标准化后得到的数据，通常也称为z-分数（z-score ）。某班级的数学考试成绩服从正态分布，已知班级的平均分为75，方差为64，小明的数学考试得分为85。请计算小明的 z-score（标准分数）（保留小数点后2位）。 （0.5 point）

答：

数学成绩总体M服从正态分布：$ M \sim N(\mu,\sigma^2) $ ，其中 $ \mu=75, \sigma ^2=64 $.

若将该正态分布标准化，则 $ x' = \frac{x-\mu}{\sigma} $.

小明的原始分 $x_0=85$，则标准化后 $ x_Z = \frac{85-75}{\sqrt{64}}=1.25 $ .

## 第三题

(1.0 point) 已知样本数据：[25, 30, 35, 40, 45, 30, 50, 55, 60, 65, 70, 75, 30, 80, 85, 60, 55, 70]，请用Python函数：

- 计算该组数据的均值、中位数、众数、方差、标准差和标准误；

- 计算峰度、偏度、变异系数；

- 画出箱体图，并说明箱体图中的5个统计量。

```python
#创建数组，储存样本数据
sample=np.array([25, 30, 35, 40, 45, 30, 50, 55, 60, 65, 70, 75, 30, 80, 85, 60, 55, 70])
#计算各种统计量
print("样本数据的均值是：",round(sample.mean(),2))
print("中位数是：",sta.median(sample))
print("众数是：",sta.mode(sample))
print("方差是：",sta.variance(sample))
print("标准差是：",round(sta.stdev(sample), 2))
print("标准误是：",round(stats.sem(sample), 2))
print("峰度是：",round(stats.kurtosis(sample), 2))
print("偏度是：",round(stats.skew(sample),2))
print("变异系数是：",round(stats.variation(sample),2))

# 绘制箱型图
sns.boxplot(y=sample, color="lightblue")

# 在图上标注文字
plt.text(0,np.median(sample),"中位数",ha="center",color="green", fontsize=12)
plt.text(0,np.percentile(sample,25),"Q1:25分位数",ha="center",color="blue")
plt.text(0,np.percentile(sample,75),"Q3:75分位数",ha="center",color="blue")
plt.text(0,np.min(sample),"最小值",ha="center",color="red")
plt.text(0,np.max(sample),"最大值",ha="center",color="red")
plt.title("样本数据箱型图")
plt.ylabel("数值")
plt.show()
```

需要注意的是，这里的最大、最小值，均指的是非异常范围内的最值。

我们定义 $IQR=Q3-Q1$ ，最大值 $max=Q3+1.5 \times IQR $  ，最小值 $ min=Q1-1.5 \times IQR $ .
