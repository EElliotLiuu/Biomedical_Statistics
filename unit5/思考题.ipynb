## 思考题 1   ·   编程练习
某团队测试两种降压药 (Drug X & Drug Y):  

 X 组 200 人中 156 人血压恢复正常;   Y 组 180 人中 126 人恢复正常。
 
请计算两组的 95% CI 与差异 CI,  并执行 z -检验。

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import statistics as sta
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

from statsmodels.stats.proportion import proportion_confint
m_x, n_x, alpha=156, 200, 0.05
l_x, h_x=proportion_confint(m_x,n_x,alpha,method="normal")
print("x组血压恢复正常率的95%置信区间是 : [{:.2f}, {:.2f}]".format(l_x, h_x))
m_y, n_y=126, 200
l_y, h_y=proportion_confint(m_y,n_y,alpha,method="normal")
print("y组血压恢复正常率的95%置信区间是 : [{:.2f}, {:.2f}]".format(l_y, h_y))
```

```python
from statsmodels.stats.proportion import confint_proportions_2indep
lo, hi = confint_proportions_2indep(
        156,200,126,180,
        method="wald", compare="diff")
print("量组血压恢复正常率的差值的95%置信区间是 : [{:.2f}, {:.2f}]".format(lo, hi))
```

```python
from statsmodels.stats.proportion import proportions_ztest

count = np.array([156, 126])
nobs  = np.array([200, 180])
z, p = proportions_ztest(count, nobs)
print('z统计量为', z.round(2))
print('p值为', p.round(2))
```

置信区间包括0，说明有95%的信心两组无差异。

$p=0.08>\alpha=0.05$，我们不能拒绝零假设：两组正常率相等。

## 思考题 2   ·   两家餐馆的好评率
两家同类菜系的餐馆好评信息:    

·  新店:   50 个评价中 40 个点赞

·  老店:  400 个评价中 300 个点赞
 
能否仅根据两个 95% CI 是否重叠,  判断两家好评率有无差异?  请编程求解并解释。

```python
c_n, n_n, c_o, n_o= 40, 50, 300, 400
alpha=0.05

l_x, h_x=proportion_confint(c_n,n_n,alpha,method="normal")
print("新店好评率的95%置信区间是 : [{:.2f}, {:.2f}]".format(l_x, h_x))
m_y, n_y=126, 200
l_y, h_y=proportion_confint(c_o,n_o,alpha,method="normal")
print("老店好评率的95%置信区间是 : [{:.2f}, {:.2f}]".format(l_y, h_y))
```

```python
from statsmodels.stats.proportion import confint_proportions_2indep
lo, hi = confint_proportions_2indep(
        40, 50, 300, 400,
        method="wald", compare="diff")
print("两店好评率差值的95%置信区间是 : [{:.2f}, {:.2f}]".format(lo, hi))
```

```python
from statsmodels.stats.proportion import proportions_ztest

count = np.array([40,300])
nobs  = np.array([50,400])
z, p = proportions_ztest(count, nobs)
print('z统计量为', z.round(2))
print('p值为', p.round(2))
```

## 课后练习 1  ·  独立性
某校 600 学生  ·  性别 × 近视：
 男 180/320，女 165/280

请用 chi2_contingency 检验性别与近视是否独立，报告 χ²、df、p、V。

```python
from scipy.stats import chi2_contingency, chisquare
table = np.array([
    [180, 320-180], 
    [165, 280-165]  
])
chi2, p, df, expected = chi2_contingency(table, correction=False)
N = table.sum()
k = min(table.shape)
V = np.sqrt(chi2 / (N * (k - 1)))
print("\n期望频数：")
print(expected)
print("最小期望频数：", np.min(expected))
print("所有期望频数≥5？", np.all(expected >= 5))

print(f"\nχ²({df}, N = {N}) = {chi2:.2f}, p = {p:.3f}, Cramér's V = {V:.2f}")
```
