# Part 3: Dispersion  & Distribution
## 函数1：import statistics as sta 

* sta.pvariance
* sta.variance
* sta.pstdev
* sta.stdev

## 函数2： import numpy as np

* np.var(ddof=0, or 1) 
* np.std(ddof=0, or 1)

    ddof=1，则分母为N-1


```python
A = np.array([10,14,11,7,9.5,15,19])
print("The population variance of ",A, "sta.pvariance is", sta.pvariance(A).round(2))
print("The sample variance of ",A, "sta.variance is", sta.variance(A).round(2))
print("The sample variance of ",A, "by np.var is", np.var(A,ddof=1).round(2))

print("The population standarde deviation of ",A, "is", np.round(sta.pstdev(A),2))
print("The sample standard deviation of ",A, "is", np.round(sta.stdev(A),2))
print("The sample standard deviation of ",A, "is", np.std(A,ddof=1).round(2))

```

# Part 4: Distribution

## 函数： import scipy.stats as stats

* stats.skew
* stats.kurtosis

```python
A = np.array([10,14,11,7,9.5,15,19])
print("The skewness of ",A, "is", np.round(stats.skew(A),2))
print("The kurtosis of ",A, "is", np.round(stats.kurtosis(A,fisher=True),2))
print("The kurtosis of ",A, "is", np.round(stats.kurtosis(A,fisher=False),2))

print("The sample coefficient variation of ",A, "is", np.round(sta.stdev(A)/sta.mean(A),2))
print("The poppulation coefficient variation of ",A, "is", np.round(sta.pstdev(A)/sta.mean(A),2))

print("The  coefficient variation of ",A, " by stats.variation is", stats.variation(A).round(2))
```

# 课堂练习：
1. 针对mtcars数据

* vs, am , gear, carb 的Mode
* cyl 的high, low median
* mpg的range, IQR, mean, sample variance, sample standard deviation, population cv
* wt的skewness, kurtosis ,population variation, 结合第一单元所学的qqplot，评价wt分布的正态性


```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import statistics as sta
import statsmodels.api as sm
import matplotlib.pyplot as plt


data=pd.read_csv("mtcars.csv")
data.head()
print("众数是：","\n" , data[["cyl", "am", "gear", "carb"]].mode() )
print("cyl的high、low median分别是：", sta.median_high(data["cyl"]), sta.median_low(data["cyl"]))
print("mpg的range, IQR, mean, sample variance, sample standard deviation, population cv分别是：",
      np.max(data["mpg"]) - np.min(data["mpg"]),np.percentile(data["mpg"],75)-np.percentile(data["mpg"],25),data["mpg"].mean(),np.var(data["mpg"],ddof=1),np.std(data["mpg"],ddof=1),
      stats.variation(data["mpg"]))
print("wt的skewness, kurtosis ,population variation分别是：",stats.skew(data["wt"]),stats.kurtosis(data["wt"]),stats.variation(data["wt"]))

fig = sm.qqplot(data["wt"], stats.norm,line='s')
plt.show()
```
